```python
# Модуль 1, загрузка данных
import pandas as pd
import numpy as np
import os
from datetime import datetime

# Конфигурация
DATA_PATH = 'data/'

# Загрузка всех файлов
bank_clients = pd.read_csv(DATA_PATH + 'bank_clients.tsv', sep='\t')
bank_complaints = pd.read_csv(DATA_PATH + 'bank_complaints.tsv', sep='\t')
bank_transactions = pd.read_csv(DATA_PATH + 'bank_transactions.tsv', sep='\t')
ecosystem_mapping = pd.read_csv(DATA_PATH + 'ecosystem_mapping.tsv', sep='\t')
market_delivery = pd.read_csv(DATA_PATH + 'market_place_delivery.tsv', sep='\t')
mobile_build = pd.read_csv(DATA_PATH + 'mobile_build.tsv', sep='\t')
mobile_clients = pd.read_csv(DATA_PATH + 'mobile_clients.tsv', sep='\t')
```

```python
# модуль 2, поиск мошенников по жалобе

# Алгоритм:
# 1. Для каждой жалобы извлекаем сумму из текста
# 2. Находим транзакцию жертвы на эту сумму
# 3. Определяем счет получателя (мошенника)
# 4. Ищем владельца счета или связанные данные

import re

def extract_amount_from_complaint(text):
    """
    Извлекает сумму из текста жалобы
    """
    patterns = [
        r'(\d+)[\s]?р',           # 15000р
        r'(\d+)[\s]?руб',          # 15000руб
        r'(\d+)[\s]?рублей',       # 15000 рублей
        r'(\d+)[\s]?₽',            # 15000₽
        r'(\d+)[\s]?[\.,]?[\s]?р', # 15000 р. или 15000,р
    ]
    
    for pattern in patterns:
        match = re.search(pattern, text.lower())  # lower() для обработки "Р" и "р"
        if match:
            return int(match.group(1))
    
    # Если не нашли по паттернам, пробуем найти просто число после слова "пропали"
    # Это запасной вариант
    if 'пропали' in text.lower():
        words = text.split()
        for i, word in enumerate(words):
            if word.lower() == 'пропали' and i + 1 < len(words):
                # Пробуем извлечь число из следующего слова
                next_word = words[i + 1]
                numbers = re.findall(r'\d+', next_word)
                if numbers:
                    return int(numbers[0])
    
    return None  # Если сумма не найдена

def find_fraud_by_complaint(complaint_row, bank_clients_df, bank_transactions_df):
    """
    Возвращает словарь с данными или None, если мошенник не найден.
    """
    # Извлекаем данные жалобы
    victim_id = complaint_row['uerId']
    complaint_text = complaint_row['text']
    complaint_date = complaint_row['event_date']
    
    # получаем сумму из текста
    amount = extract_amount_from_complaint(complaint_text)
    if not amount:
        # Если сумма не извлеклась, пропускаем жалобу
        return None
    
    # Находим счет жертвы по ее ID
    victim_data = bank_clients_df[bank_clients_df['userId'] == victim_id]
    if victim_data.empty:
        # Жертва не найдена в базе клиентов
        return None
    
    victim_account = victim_data.iloc[0]['accout']
    victim_phone = victim_data.iloc[0]['phone']
    
    # транзакции на конкретную сумму по жалобам
    fraud_transaction = bank_transactions_df[
        (bank_transactions_df['account_out'] == victim_account) & 
        (bank_transactions_df['value'] == amount)
    ]
    
    if fraud_transaction.empty:
        # Транзакция на эту сумму не найдена
        return None
    
    # Берем первую транзакцию 
    transaction = fraud_transaction.iloc[0]
    fraud_account = transaction['account_in']
    transaction_date = transaction['event_date']
    
    # владалец счета
    fraud_bank_owner = bank_clients_df[bank_clients_df['accout'] == fraud_account]
    
    # Формируем результат
    result = {
        'complaint_id': victim_id,
        'complaint_text': complaint_text,
        'complaint_date': complaint_date,
        'extracted_amount': amount,  # Сумма, которую извлекли из жалобы
        'victim_account': victim_account,
        'victim_phone': victim_phone,
        'transaction_date': transaction_date,
        'fraud_account': fraud_account,
        'fraud_bank_owner_id': fraud_bank_owner.iloc[0]['userId'] if not fraud_bank_owner.empty else None,
        'fraud_bank_owner_fio': fraud_bank_owner.iloc[0]['fio'] if not fraud_bank_owner.empty else None,
        'fraud_bank_owner_phone': fraud_bank_owner.iloc[0]['phone'] if not fraud_bank_owner.empty else None
    }
    
    return result

fraud_cases = []

for idx, complaint in bank_complaints.iterrows():
    case = find_fraud_by_complaint(complaint, bank_clients, bank_transactions)
    if case:
        fraud_cases.append(case)
    else:
        pass

fraud_cases_df = pd.DataFrame(fraud_cases)
```

```python
# поиск телефонных связей
# Для каждого найденного мошенника ищем звонки между ним и жертвой

def find_calls_between(victim_phone, fraud_account, 
                       mobile_build_df, mobile_clients_df, 
                       ecosystem_mapping_df, transaction_date):
    """
    Находит звонки между жертвой и мошенником.
    Мошенника ищем через связку счет -> bank_id -> mobile_id -> телефон
    """
    # Преобразуем телефон жертвы в числовой формат
    victim_phone_int = int(victim_phone)
    
    # Ищем владельца счета мошенника
    fraud_bank_owner = bank_clients[bank_clients['accout'] == fraud_account]
    if fraud_bank_owner.empty:
        return None
    
    fraud_bank_id = fraud_bank_owner.iloc[0]['userId']
    
    # Ищем в ecosystem_mapping связь с мобильным оператором
    mapping = ecosystem_mapping_df[ecosystem_mapping_df['bank_id'] == fraud_bank_id]
    if mapping.empty:
        return None
    
    fraud_mobile_id = mapping.iloc[0]['mobile_user_id']
    if pd.isna(fraud_mobile_id):
        return None
    
    # Ищем телефон мошенника в mobile_clients
    fraud_mobile = mobile_clients_df[mobile_clients_df['client_id'] == fraud_mobile_id]
    if fraud_mobile.empty:
        return None
    
    fraud_phone_int = int(fraud_mobile.iloc[0]['phone'])
    
    # Ищем звонки между ними
    calls = mobile_build_df[
        ((mobile_build_df['from_call'] == victim_phone_int) & (mobile_build_df['to_call'] == fraud_phone_int)) |
        ((mobile_build_df['from_call'] == fraud_phone_int) & (mobile_build_df['to_call'] == victim_phone_int))
    ]
    
    if calls.empty:
        return None
    
    # Добавляем информацию о мошеннике
    result = calls.copy()
    result['fraud_phone'] = fraud_phone_int
    result['fraud_mobile_id'] = fraud_mobile_id
    result['fraud_fio'] = fraud_mobile.iloc[0]['fio']
    
    return result

# Применяем ко всем найденным случаям
for idx, case in fraud_cases_df.iterrows():
    calls = find_calls_between(
        case['victim_phone'], 
        case['fraud_account'],
        mobile_build, 
        mobile_clients, 
        ecosystem_mapping,
        case['transaction_date']
    )
    # Здесь можно сохранять звонки для каждого случая
    # Для простоты добавим флаг, были ли звонки
    fraud_cases_df.loc[idx, 'has_calls'] = 0 if calls is None else 1
```

```python
# модуль 4, маркетплейсы
# Для каждого мошенника ищем его активность в маркетплейсе

def find_market_activity(fraud_account, bank_clients_df, ecosystem_mapping_df, market_delivery_df):
    """
    Находит доставки мошенника в маркетплейсе
    """
    # Ищем владельца счета
    fraud_bank_owner = bank_clients_df[bank_clients_df['accout'] == fraud_account]
    if fraud_bank_owner.empty:
        return None
    
    fraud_bank_id = fraud_bank_owner.iloc[0]['userId']
    
    mapping = ecosystem_mapping_df[ecosystem_mapping_df['bank_id'] == fraud_bank_id]
    if mapping.empty:
        return None
    
    fraud_market_id = mapping.iloc[0]['market_plece_user_id']
    if pd.isna(fraud_market_id):
        return None
    
    # Ищем доставки
    deliveries = market_delivery_df[market_delivery_df['user_id'] == fraud_market_id]
    
    if deliveries.empty:
        return None
    
    result = deliveries.copy()
    result['fraud_bank_id'] = fraud_bank_id
    result['fraud_market_id'] = fraud_market_id
    
    return result

# %%
# Применяем ко всем найденным случаям
for idx, case in fraud_cases_df.iterrows():
    market_data = find_market_activity(
        case['fraud_account'],
        bank_clients,
        ecosystem_mapping,
        market_delivery
    )
    fraud_cases_df.loc[idx, 'has_market_activity'] = 0 if market_data is None else 1
    if market_data is not None:
        fraud_cases_df.loc[idx, 'market_deliveries_count'] = len(market_data)
```

```python
# модуль 5, сохранение
# Сохраняем все найденные случаи в CSV для дальнейшего использования

fraud_cases_df.to_csv(DATA_PATH + 'fraud_cases_detected.csv', index=False)

# Для каждого случая с активностью сохраняем детали
for idx, case in fraud_cases_df.iterrows():
    if case['has_market_activity']:
        market_data = find_market_activity(
            case['fraud_account'],
            bank_clients,
            ecosystem_mapping,
            market_delivery
        )
        if market_data is not None:
            filename = f"fraud_market_{case['fraud_account']}.csv"
            market_data.to_csv(DATA_PATH + filename, index=False)
    
    if case['has_calls']:
        calls = find_calls_between(
            case['victim_phone'], 
            case['fraud_account'],
            mobile_build, 
            mobile_clients, 
            ecosystem_mapping,
            case['transaction_date']
        )
        if calls is not None:
            filename = f"fraud_calls_{case['fraud_account']}.csv"
            calls.to_csv(DATA_PATH + filename, index=False)
```

```python
# Формируем структуру для загрузки в графовую БД

def prepare_neo4j_data(fraud_cases_df, bank_clients_df, mobile_clients_df, ecosystem_mapping_df):
    """
    Создает CSV для загрузки в Neo4j:
    - nodes: люди, счета, телефоны
    - edges: связи между ними
    """
    neo4j_nodes = []
    neo4j_edges = []
    
    for idx, case in fraud_cases_df.iterrows():
        # Узел: жертва
        neo4j_nodes.append({
            'id': f"victim_{case['complaint_id']}",
            'type': 'person',
            'role': 'victim',
            'bank_id': case['complaint_id'],
            'account': case['victim_account'],
            'phone': case['victim_phone']
        })
        
        # Узел: мошенник (если есть банковский владелец)
        if pd.notna(case['fraud_bank_owner_id']):
            neo4j_nodes.append({
                'id': f"fraud_{case['fraud_bank_owner_id']}",
                'type': 'person',
                'role': 'fraud',
                'bank_id': case['fraud_bank_owner_id'],
                'account': case['fraud_account'],
                'phone': case['fraud_bank_owner_phone'],
                'fio': case['fraud_bank_owner_fio']
            })
            
            # Связь: перевод денег
            neo4j_edges.append({
                'from': f"victim_{case['complaint_id']}",
                'to': f"fraud_{case['fraud_bank_owner_id']}",
                'type': 'TRANSFERRED',
                'amount': case['amount'],
                'date': case['transaction_date']
            })
    
    return pd.DataFrame(neo4j_nodes), pd.DataFrame(neo4j_edges)

nodes_df, edges_df = prepare_neo4j_data(fraud_cases_df, bank_clients, mobile_clients, ecosystem_mapping)

nodes_df.to_csv(DATA_PATH + 'neo4j_nodes.csv', index=False)
edges_df.to_csv(DATA_PATH + 'neo4j_edges.csv', index=False)
```
